## W5A1 Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> 

```python
def max_of_two(a,b):
    if a<b:
        return b
    elif a>b:
        return a
    else:
        return "a=b"
a, b = map(int, input("nhập 2 số nguyên: ").split())
max_of_two(a, b)
```
