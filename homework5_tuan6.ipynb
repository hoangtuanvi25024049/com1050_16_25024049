## W5A1 Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->

## W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 


```python
def prime_checking(n):
    dem = 0
    for i in range(2,n):
        if n%i != 0:
            dem += 1
    else:
        if dem == n-2:
            print(f"{n} là số nguyên tố")
        else:
            print(f"{n} không phải số nguyên tố")
n = int(input("nhập n:"))
prime_checking(n)
```

## W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def perf_number(n):
    ds = []
    for i in range(1,n):
        if n%i == 0:
            ds.append(i)
    else:
        if sum(ds) == n:
            print(f"{n} là số hoàn hảo")
        else:
            print(f"{n} không phải số hoàn hảo")
n = int(input("nhập n:"))
perf_number(n)
```

## W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử, đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def find_number(ds,k):
    for i in ds:
        if i == k:
            print(ds.index(i)+1)
            break
        else:
            continue
    else:
        print("-1")
ds = list(map(int,input("nhập 1 dãy số nguyên:").split()))
k = int(input("nhập 1 số nguyên:"))
find_number(ds,k)
```

## W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    tich = 1
    for i in range(1,n+1):
        tich *= i
    else:
        print(tich)

n = int(input("nhập n:"))
giai_thua(n)
```

## W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
## Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def hamming_dst(a,b):
    str_a = len(str(format(a,'b')))
    str_b = len(str(format(b,'b')))
    if str_a > str_b:
        print(f"khoảng cách hamming của 2 số là: {str_a - str_b}")
    elif str_a < str_b:
        print(f"khoảng cách hamming của 2 số là: {str_b - str_a}")
    else:
        print(f"khoảng cách hamming của 2 số là: {str_a - str_b}")
a,b = map(int,input("nhập 2 số:").split())
hamming_dst(a,b)
```

## W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def tong(n):
    chu_so = []
    while n != 0:
        chu_so.append(n%10)
        n = n//10
    print(f"tổng các chữ số trong số đã nhập là: {sum(chu_so)}")
n = int(input("nhập n:"))
tong(n)
```
