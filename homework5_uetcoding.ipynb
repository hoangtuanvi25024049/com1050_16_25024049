## Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
n = int(input())
if n < 2:
  print("no")
elif n == 2:
  print("2")
else:
  dem = 0
  ds = []
  for i in range(2,n+1):
    if n%i == 0:
      for k in range(2,int(i**0.5)+1):
        if i%k == 0:
          dem += 1
          break 
      if dem == 0:
         ds.append(i)
        
print(max(ds))
```

## Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.



```python
n = input()
dem = 0
while True:
    if n[::-1] == n:
        print(dem, n)
        break
    else:
        n = int(n) + int(n[::-1])
        n = str(n)
        dem += 1
    if dem > 100:
        print("stop")
        break
```

## Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).



```python
n = int(input("nhập số nguyên dương:"))
ds = []
if n > 1:
    ds.append(1)
for i in range(2,n+1):
    for j in range(2,i+1):
        if (j**0.5)%1 == 0:
            if j not in ds:
                ds.append(j)
            else:
                continue
if ds == []:
    print("(no number)")
else:
    print(*ds)
        
```

```python
n = int(input("nhập số nguyên dương:"))
ds = []
lst_so = []
for i in range(1,n+1):
    if i**2 <= n:
        ds.append(i**2)
    else:
        break

for i in ds:
    if len(set(str(i))) == len(str(i)):
        lst_so.append(i)
    else:
        continue

if lst_so == []:
    print("(no number)")
else:
    print(*lst_so)
```
