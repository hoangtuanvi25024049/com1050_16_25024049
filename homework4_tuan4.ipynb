##  Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b = map(int, input("nhập hai số nguyên: ").split())
if a>b:
    for i in range(1,b+1):
        if a%i==0 and b%i==0:
            print(i)
elif a<b:
    for i in range(1,a+1):
        if a%i==0 and b%i==0:
            print(i)
else:
    print(a)
```

## Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input("nhập một số nguyên: "))
a = n
dem = 0
while n%2 == 0:
    n = n/2
    dem += 1
    if n%2 != 0:
        print(f"{a} = 2^{dem}")
        break
else:
    print("số bạn nhập không có dạng 2^k")
```

## Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("nhập một số nguyên dương: "))
tong = 0
while n>0:
    tong += n%10
    n = n//10
print(f"tổng các chữ số của số bạn nhập là: {tong}")
```

## Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("nhập một số nguyên dương: "))
so_chan = 0
so_le = 0
while n>0:
    a = n%10
    n = n//10
    if a%2 == 0:
        so_chan += 1
    else:
        so_le += 1
print(f"số lượng chữ số chẵn: {so_chan}")
print(f"số lượng chữ số lẻ: {so_le}")
```

## Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
n = int(input("nhập một số nguyên: "))
tong = 0
i = 1
while True:
    tong = tong + 1/i
    i = i + 1
    if tong > n:
        print(f"số k nhỏ nhất để S(k) > {n} là: {i-1}")
        break
```

## Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
kt = []
while True:
    n = int(input("nhập một số nguyên (nhập -1 để dừng): "))
    if n != -1:
        kt.append(n)
    else:
        break
print(f"số lớn nhất trong dãy là: {max(kt)}")
print(f"số nhỏ nhất trong dãy là: {min(kt)}")
```

## Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
n = int(input("nhập một số nguyên: "))
so_thu1 = 0
so_thu2 = 1
while True:
    if so_thu2 < n:
        so_thu1, so_thu2 = so_thu2, so_thu1 + so_thu2
        if (so_thu2 > n) and (so_thu1 < n):
            print(f"{so_thu1} là số Fibonacci lớn nhất nhỏ hơn {n}")
            break
```

## Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng). Dùng cách: for trên index; KHÔNG dùng hàm     split(), len()

```python
chuoi = input("nhập một chuỗi: ")
dem = 0
while True:
    if chuoi == "":
        print("bạn đã nhập chuỗi rỗng, vui lòng nhập lại")
        chuoi = input("nhập một chuỗi: ")
    else:
        for i in chuoi:
            if i.isalpha() == False:
                dem += 1
        else:
            if i.isalpha() == True:
                dem += 1
                break
print(f"số từ trong chuỗi là: {dem}")
```

##  Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi. Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
n = input("nhập 1 chuỗi:")
ds = []
for i in n:
    if i.isalpha() == True:
        ds.append(i)
    else:
        break
print(f"từ đầu tiên của chuỗi là: {''.join(ds)}") 
```

## Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. 

```python
n = list(input("nhập một chuỗi: "))
for i in range(len(n)):
    if n[i].isupper() == True:
        n[i] = n[i].lower()
    elif n[i].islower() == True:
        n[i] = n[i].upper()
else:
    print(f"chuỗi sau khi đổi: {''.join(n)}")
```

## Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số 

```python
n = input("nhập 1 chuỗi:")
dct = 0
dch = 0
dem_so = 0
for i in n:
    if i.isalpha() == True:
        if i.isupper() == True:
            dch += 1
        else:
            dct += 1
    elif i.isdigit() == True:
        dem_so += 1
print(f"số chữ hoa: {dch}")
print(f"số chữ thường: {dct}")
print(f"số chữ số: {dem_so}")
```
