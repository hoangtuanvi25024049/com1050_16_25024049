## Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h+1):
    for j in range(1, 2*h):
        if (j == h-i+1) or (j == h+i-1) or (i == h):
            print("*", end="")
        else:
            print(" ", end="")
    print() 
```

## Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a

```python
n = int(input("nhập 1 số nguyên dương:"))
ds = []
if n>0:
    for i in range(1,n+1):
        if n%i==0:
            ds.append(i)
    print("Các ước số của", n, "là:", " ".join(map(str, ds)))
else:
    print("không phải số nguyên dương")
```

## Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a,b = map(int, input("nhập 2 số nguyên dương a và b: ").split())
ds = []
for i in range(1, min(a,b)+1):
    if a%i==0 and b%i==0:
        ds.append(i)
print(f"Ước chung của {a} và {b} là: {" ".join(map(str, ds))}")
```

## Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 


```python
n = int(input("Nhập một số nguyên dương: "))
if n>0:
    for i in range(1,n+1):
        if n/i==i:
            print(f"{n} là số chính phương")
            break
        else:
            print(f"{n} không phải là số chính phương")
            break
else:
    print("không phải số nguyên dương")
```

## Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
n = int(input("Nhập một số nguyên dương: "))
ds = []
for i in range(1,n):
    if n%i == 0:
        ds.append(i)
if sum(ds) == n:
    print(f"{n} là số hoàn hảo")
else:
    print(f"{n} không phải là số hoàn hảo")

```

## Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
a = int(input("Nhập số nguyên dương a: "))
tong = 0
n = 0
while tong <= a:
    n += 1
    tong += 1/n
print(f"số nguyên dương n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n > {a} là: {n}")
```

## Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
_min = None
_max = None
## khởi tạo biến None tức biến rỗng để kiểm tra giá trị nhập vào
while True:
    n = int(input("Nhập một số nguyên dương (nhập -1 để dừng): "))
    if n == -1:
        break
    if _min is None or n < _min: 
        ## nếu _min chưa có giá trị hoặc n nhỏ hơn _min thì cập nhật _min
        _min = n
    if _max is None or n > _max: 
        ## nếu _max chưa có giá trị hoặc n lớn hơn _max thì cập nhật _max
        _max = n
if _min is None and _max is None: 
    ## kiểm tra nếu người dùng không nhập số nào ngoài -1
    print("Bạn không nhập số nào.")
elif _min == _max: 
    ## kiểm tra nếu số nhỏ nhất và số lớn nhất bằng nhau (tức là người dùng chỉ nhập một số duy nhất)
    print(f"số nhỏ nhất và số lớn nhất bạn đã nhập đều là: {_min}")
else:
    print(f"Số nhỏ nhất bạn đã nhập là: {_min}, số lớn nhất bạn đã nhập là: {_max}")
```

## Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng. 

```python
n = int(input("Nhập một số nguyên dương: "))
dem = 0
while True:
    n = n//10
    dem += 1
    if n == 0:
        break
print(f"số bạn nhập có {dem} chữ số")

```

```python
## cách khác đếm số chữ số bằng cách chuyển số thành chuỗi và đếm độ dài của chuỗi theo AI
def dem_chu_so(n):
    if n < 10:
        return 1
    return 1 + dem_chu_so(n // 10)

## ví dụ với số 999 hàm thực hiện trả về 1 trước và thực hiện việc //10 giảm số xuống 99
## sau đó tiếp tục trả về 1 và //10 giảm số xuống 9
## tiếp tục trả về 1 và //10 giảm số xuống 0, lúc này n < 10 nên trả về 1, tổng cộng là 3 chữ số của 999

n = int(input("Nhập n: "))
print(f"Số chữ số là: {dem_chu_so(abs(n))}")
```

## Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input("Nhập một số nguyên dương: "))
dem = 0
while True:
    chu_so_cuoi = (n%10)%2 
    # gán biến kiểm tra chữ số cuối

    if chu_so_cuoi != 0:
        dem += 1
    n = n//10
    # bỏ chữ số cuối, xét tiếp các số phía trước

    if n==0:
        break 
    # kiểm tra sau các bước trên nếu số n còn lại = 0 thì dừng

print(f"số bạn nhập có {dem} chữ số lẻ")
```

## Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

## Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 


```python
n = int(input("Nhập một số nguyên dương: "))
a = n
dem = 0
while True:
    phan_du = n%3
    if phan_du == 0:
        n = n//3
        dem += 1
    if phan_du == 1 or phan_du == 2:
        break
if a == 3**dem:
    print(f"số {a} có dạng 3^{dem} ")
else:
    print(f"số {a} không có dạng 3^k với k là số nguyên dương")
```

## Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
n = int(input("Nhập một số nguyên dương: "))
a = 0
b = 1
while b <= n:
    o_v = b
    b = a + b
    a = o_v
print(f"số Fibonacci lớn nhất không vượt quá {n} là: {a}")
    
  
```

## Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input("Nhập một số nguyên dương: "))
a = n
tong = 0
while n > 0:
    phan_du = n%10
    if phan_du in [1, 2, 3, 5, 7]:
        tong += 1
    n = n // 10
print(f"Số chữ số là số nguyên tố trong {a} là: {tong}")
```
