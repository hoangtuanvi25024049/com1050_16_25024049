## nhập vào cân nặng và chiều cao của 2 người, in ra bmi cao hơn

```python
def compare_(w1, h1, w2, h2):
    bmi1 = w1/(h1**2)
    bmi2 = w2 / (h2**2)
    if bmi1 > bmi2:
        return 1
    elif bmi1 < bmi2:
        return 2
    else:
        return 0
w1, h1 = map(float,input("nhập cân nặng và chiều cao người thứ 1: ").split())
w2, h2 = map(float,input("nhập cân nặng và chiều cao người thứ 2: ").split())
if compare_(w1, h1, w2, h2) == 1:
    print("người 1 có bmi cao hơn")
elif compare_(w1, h1, w2, h2) == 2:
    print("người 2 có bmi cao hơn")
else:
    print("hai người có bmi bằng nhau")
```

## given a score s, return a letter grade bt the rules:
<!-- A+ : s >= 9
A  : s >= 8.5
B+ : S >=8
B  : s >= 7
E  : otherwise -->

```python
def grade(s):
    if s >= 9:
        return "A+"
    elif s >= 8.5:
        return "A"
    elif s >= 8:
        return "B+"
    elif s >= 7:
        return "B"
    else:
        return "E"
s = float(input("nhập điểm số:"))
print(grade(s))
```

## given an integer n, return:
5*n + 1 if n mod 3 == 1
6*n + 2 if n mod 3 == 2
7*n + 3 if n mod 3 == 3

```python
def phandu(x):
    if x%3 == 1:
        return 5*x+1
    elif x%3 == 2:
        return 6*x+2
    else:
        return 7*x+3
phandu(int(input("nhập 1 số nguyên:")))
```

## bài toán tìm số nguyên tố

```python
ds = []
def prime_checking(a,b):
  for n in range(a,b+1):
    if n<2:
      continue
    if n==2:
      ds.append(2)
      continue
    dem = 0
    for i in range(2,int(n**0.5)+1):
      if n%i == 0:
        dem += 1
        break
    if dem == 0:
        ds.append(n)
    
a,b = map(int,input("nhập 2 số nguyên:").split())
prime_checking(a,b)
print(ds)

```
