Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c = map(int,input("nhập 3 cạnh của tam giác:").split())
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("diện tích, chu vi của tam giác đều lần lượt là:", round((((a**2 -a**2/4)**0.5)*a)/2, 2), a*3)
    elif a==b or a==c or b==c:
        if a==b:
            h = ((a**2) - (c**2/4))**0.5
            print("diện tích, chu vi của tam giác cân lần lượt là:", round((h*c)/2, 2), a*2+c)
        elif a==c:
            h = ((a**2) - (b**2/4))**0.5
            print("diện tích, chu vi của tam giác cân lần lượt là:", round((h*b)/2, 2), a*2+b)
        else:
            h = ((b**2) - (a**2/4))**0.5
            print("diện tích, chu vi của tam giác cân lần lượt là:", round((h*a)/2, 2), b*2+a)
    elif (a**2 == b**2 + c**2) or (b**2 == a**2 + c**2) or (c**2 == a**2 + b**2):
        if a**2 == b**2 + c**2:
            print("diện tích, chu vi của tam giác vuông lần lượt là:", round((b*c)/2, 2), a+b+c)
        elif b**2 == a**2 + c**2:
            print("diện tích, chu vi của tam giác vuông lần lượt là:", round((a*c)/2, 2), a+b+c)
        else:
            print("diện tích, chu vi của tam giác vuông lần lượt là:", round((a*b)/2, 2), a+b+c)    
    else:
        p = (a+b+c)/2
        print("diện tích, chu vi của tam giác thường lần lượt là:", round((p*(p-a)*(p-b)*(p-c))**0.5, 2), a+b+c)
else:
    print("không phải là tam giác")
```

## Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0)

```python
print("phương trình bậc nhất có dạng ax + b = 0 với a khác 0")
a,b = map(float,input("nhập a và b:").split())
if a != 0:
    print("nghiệm của phương trình là:", round(-b/a, 2))
else:
    print("không giải được phương trình bậc nhất vì a bằng 0")
```

## Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
print("phương trình bậc 2 có dạng ax^2 + bx + c = 0 với a khác 0")
a,b,c = map(float,input("nhập a, b và c:").split())
if a != 0:
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = round((-b+delta**0.5)/(2*a), 2)
        x2 = round((-b-delta**0.5)/(2*a), 2)
        print("phương trình có hai nghiệm phân biệt là:", x1, "và", x2)
    elif delta == 0:
        x = round(-b/(2*a), 2)
        print("phương trình có nghiệm kép là:", x)
    else:
        print("phương trình vô nghiệm vì delta âm")
else:
    print("không phải là phương trình bậc 2 vì a bằng 0")
```

## Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

```python
a,b,c,d = map(int,input("nhập a, b, c và d:").split())
list = [a,b,c,d]
_max = a
for i in range(1,4):
    if list[i] > _max:
        _max = list[i]
print("giá trị lớn nhất trong 4 số là:", _max)
```

## Bài 8: Giải hệ phương trình tuyến tính:
## ax + by = m
## cx + dy = n

```python
print("hệ phương trình tuyến tính có dạng:")
print(" a1*x + b1*y = c1","\n","a2*x + b2*y = c2")
a1,b1,c1,a2,b2,c2 = map(float,input("nhập a1, b1, c1, a2, b2 và c2:").split())
D = a1*b2 - a2*b1
Dx = c1*b2 - c2*b1
Dy = a1*c2 - a2*c1
if D != 0:
    x = round(Dx/D, 2)
    y = round(Dy/D, 2)
    print("hệ phương trình có nghiệm duy nhất là: x =", x, "và y =", y)
else:
    print("hệ phương trình không có nghiệm duy nhất")
```

## Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x,y = map(int,input("nhập tọa độ tâm đường tròn:").split())
r = int(input("nhập bán kính đường tròn:"))
a,b = map(int,input("nhập tọa độ điểm cần kiểm tra:").split())
if (a-x)**2 + (b-y)**2 < r**2:
    print("điểm nằm trong đường tròn")
elif (a-x)**2 + (b-y)**2 == r**2:
    print("điểm nằm trên đường tròn")
else:
    print("điểm nằm ngoài đường tròn")
```
