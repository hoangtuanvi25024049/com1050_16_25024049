### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 
0 1

```python
lst = [10, 0, 0, 12, 34, 5, 6, 7, 8, 89, 0, 1]
def min_and_pos(lst):
    if lst:
        min_val = lst[0]
        min_pos = 1
        for i, x in enumerate(lst,1):
            if x < min_val:
                min_val = x
                min_pos = i
        return min_val, min_pos
    
# import random
# lst = [random.randrange(189,500) for _ in range(10)]
# print(lst)
# có thể dùng random để kiểm tra

assert min_and_pos(lst) == (0, 2)
print("Pass all tests")
```

### Bài 3: Tìm giá trị trung bình của các số nguyên dương của 1 list
nếu không có số dương nào trả về invalid

```python
# lst = list(map(int,input("nhập 1 danh sách số nguyên:").split()))
def aver(lst):
    if lst:
        for i in lst:
            posi_nums = []
            if i > 0:
                if i%2 == 0:
                    posi_nums.append(i)
        if not posi_nums:
            return "invalid"
        else:
            t = 0
            for x in posi_nums:
                t += x
            else:
                return t/len(posi_nums)

assert aver([2,2,2,2]) == 2
assert aver([-2, -1, -20, -343]) == "invalid"

print("pass all tests")
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không
VD: [1, 2, 3, 5, 3, 2, 1]

```python
# lst = [1, 2, 3, 5, 3, 2, 1]
def palindrome(lst):
    if lst:
        for i in range(len(lst)//2):
            if lst[i] != lst[len(lst)-1-i]:
                return False
        return True
    else:
        return None

assert palindrome([1, 2, 3, 4, 3, 2, 1]) == True
assert palindrome([1, 2, 3, 4, 3, 4, 1]) == False
assert palindrome([]) == None
print("Accepted")
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
def unique_arr(lst):
    if lst:
        ans = []
        for x in lst:
            if x not in ans:
                ans.append(x)
        return ans
    return None
assert unique_arr([1,2,3,4,3,4,2,1,3,4,2,1]) == [1,2,3,4]
print("pass")
```
