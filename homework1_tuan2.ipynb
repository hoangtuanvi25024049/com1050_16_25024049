## Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a,b = map(int,input("nhập lần lượt chiều dài và chiều rộng của mảnh đất: ").split(" "))
Ac = (b/2)**2*3.14
Ar = a*b
A = Ar - Ac
print(f"diện tích phần đất trống là: {round(A, 2)}")
```

## Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma) 

```python
n = input("nhập 1 kí tự: ")
while len(n) != 1:
    n = input("vui lòng nhập lại 1 kí tự: ")
if n.isupper() == True:
    print(f"ký tự sau khi đổi: {n.lower()}")
else:
    print(f"ký tự sau khi đổi: {n.upper()}")
```

## Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet. 


```python
n = input("nhập 1 kí tự:")
while len(n) != 1:
    n = input("vui lòng nhập lại 1 kí tự: ")
if n.isalpha() == True:
    print(f"{n} là kí tự alphabet")
else:    
    print(f"{n} không phải là kí tự alphabet")
```

## Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 


```python
n = input("nhập 1 chữ cái in hoa:")
while len(n) != 1 or n.isupper() == False:
    n = input("vui lòng nhập lại 1 chữ cái in hoa: ")
if n == "A":
    print("không có chữ cái nào đứng trước A")
else:
    a = chr(ord(n)-1)
    print(f"chữ cái in thường đứng trước {n} là: {a.lower()}")
```

## Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a,b,c = map(int, input("nhập ba cạnh của tam giác: ").split())
if a+b>c and a+c>b and b+c>a:
    print(f"diện tích tam giác là: {round(((a+b+c)/2)*((a+b+c)/2-a)*((a+b+c)/2-b)*((a+b+c)/2-c)**0.5, 1)}")
else:
    print("ba cạnh đã nhập không phải là ba cạnh của một tam giác")
```

## Bài 7: Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và thứ chín của chuỗi này.

```python
n = input("nhập 1 chuỗi kí tự có độ dài lớn hơn hoặc bằng 20 kí tự:")
while len(n) < 20:
    n = input("vui lòng nhập lại một chuỗi kí tự có độ dài lớn hơn hoặc bằng 20 kí tự: ")
print(f"chữ cái thứ 5 của chuỗi là: {n[4]} ")
print(f"chữ cái thứ 9 của chuỗi là: {n[8]} ")
```

## Kiểm tra số chẵn lẻ: Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó là số chẵn hay số lẻ.


```python
n = int(input("nhập một số nguyên: "))
if n%2 == 0:
    print(f"{n} là số chẵn")
else:
    print(f"{n} là số lẻ")
```
