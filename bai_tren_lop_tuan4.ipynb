## nhập vào 1 số nguyên đến khi được 1 số nguyên dương thì thôi, kiểm tra xem đó có phải số nguyên tố không

```python
n = -10

while n <= 0:
    n = int(input("nhập một số nguyên dương: "))
    if n == 1:
        print("1 không phải là số nguyên tố")
    else:
        for i in range(2,int(n**0.5)+1):
            if n % i == 0:
                print(f"{n} không phải là số nguyên tố")
                break
        else:
            print(f"{n} là số nguyên tố")
                
    
```

## nhập vào a,b đến khi được a,b nguyên dương, sau đó tính tổng các số nguyên tố trong khoảng [a,b]

```python
a = -10
b = -10
while (a <= 0 or b <= 0) or (a > b):
    a,b = map(int, input("nhập hai số nguyên dương (a < b): ").split())
    tong = 0
    for i in range(a,b+1):
        if i<2:
            continue
        for j in range(2,int(i**0.5)+1):
            if i % j == 0:
                print(f"{i} không phải là số nguyên tố")
                break
        else:
            tong += i
    else:
        print(f"tổng các số nguyên tố trong khoảng từ {a} đến {b} là: {tong}")
```

## Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100.

```python
n = -10
while n <= 0 or n>100:
    n = int(input("nhập một số nguyên dương (0 < n <= 100): "))
    giai_thua = 1
    for i in range(1,n+1):
        giai_thua *= i
    else:
        print(f"{n}! = {giai_thua}")
```

## Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) của số đó. Không dùng kiểu dữ liệu string.

## Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."

```python
a = -10
b = -10
while (a <= 0 or b <= 0) or (a > b):
    a,b = map(int, input("nhập hai số nguyên dương (a < b): ").split())
    tong = 0
    for i in range(a,b+1):
        if i<2:
            continue
        for j in range(2,int(i**0.5)+1):
            if i % j == 0:
                break
        else:
            tong += i
    else:
        print(f"tổng các số nguyên tố trong khoảng từ {a} đến {b} là: {tong}")
```

## Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n

```python
n = int(input("nhập một số nguyên dương: "))
ds = []
for i in range(1,int(n**0.5)+1):
    if n % i == 0:
        ds.append(i)
print(f"ước số nguyên tố lớn nhất của {n} là: {max(ds)}")
```

```python
n = int(input("nhập một số nguyên dương (n<1000): "))
ds = []
for i in range(2,n+1):
    snt = True
    for j in range(2,int(i**0.5)+1):
        if i % j == 0:
            snt = False
            break
    if snt:
        ds.append(i)
print(f"các số nguyên tố nhỏ hơn hoặc bằng {n} là: {ds}")
```

## NHẬP VÀO SỐ NGUYÊN N, TRONG KHOẢNG 1 ĐẾN N, IN RA DÃY SỐ NGUYÊN LIÊN TIẾP DÀI NHẤT KHÔNG CHỨA SỐ NGUYÊN TỐ NÀO (DÃY NÀO TÌM THẤY ĐẦU TIÊN THÌ IN RA)

```python
n = int(input("nhập một số nguyên dương (n<1000): "))

danh_dau = [True]*(n+1)

danh_dau[0] = danh_dau[1] = False

for i in range(2,int(n**0.5)+1):
    if danh_dau[i] == True:
        for j in range(i*i,n+1,i):
            danh_dau[j] = False

ds_snt = []
for i in range(2,n+1):
    if danh_dau[i] == True:
        ds_snt.append(i)

print(f"các số nguyên tố nhỏ hơn hoặc bằng {n} là: {ds_snt}")

ket_qua = []
nhom_tam = []

for i in range(2,n+1):
    if danh_dau[i] == False:
        nhom_tam.append(i)
    else:
        if nhom_tam:
            ket_qua.append(nhom_tam)
            nhom_tam = []

ds_dem = []
for i in ket_qua:
    a = len(i)
    ds_dem.append(a)
    if a == max(ds_dem):
        print(f"nhóm có số phần tử nguyên dài nhất là: {ds_dem}")


```
