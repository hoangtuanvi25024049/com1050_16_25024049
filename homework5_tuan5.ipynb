## W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def kiemtra(x):
    chu_so = []
    a = x
    while x != 0:
        chu_so.append(x%10)
        x = x//10
    tong = 0
    for i in chu_so:
        tong += i**(len(chu_so))
    if tong == a:
        print("True")
    else:
        print("False")
x = int(input("nhập x:"))
kiemtra(x)
```

## W5A17 Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên n

```python
def kiemtra(n):
    for i in range(1,n):
        if n == i**2 :
            print("True")
            break
        else:
            continue
    else:
        print("False")
n = int(input("nhập n:"))
kiemtra(n)
```
