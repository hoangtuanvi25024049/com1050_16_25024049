## W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def kiemtra(x):
    chu_so = []
    a = x
    while x != 0:
        chu_so.append(x%10)
        x = x//10
    tong = 0
    for i in chu_so:
        tong += i**(len(chu_so))
    if tong == a:
        print("True")
    else:
        print("False")
x = int(input("nhập x:"))
kiemtra(x)
```

## W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

## W5A22 Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.


```python
def sum_halves(s):
    stn = []
    sth = []
    for i in range(3):
        stn.append(s[i])
    for k in range(3,6):
        sth.append(s[k])
    stn = int("".join(stn))
    sth = int("".join(sth))
    print(stn + sth)
sum_halves(input("nhập 1 số có 6 chữ số:"))
```

## W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a,b,c):
    if a == 0:
        return -c/b
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
            return False
        elif delta == 0:
            return -b/(2*a)
        else:
            return (-b - delta**0.5)/(2*a),(-b + delta**0.5)/(2*a)
a,b,c = map(int,input("nhập 3 số:").split())
quadratic_roots_count(a,b,c)
```

## W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    vt_AB = (x2 - x1, y2 - y1)
    vt_AC = (x3 - x1, y3 - y1)
    if (vt_AB[0]/vt_AC[0]) == (vt_AB[1]/vt_AC[1]):
        return True
    else:
        return False
x1,y1,x2,y2,x3,y3 = map(int,input("nhập các tọa độ của 3 điểm:").split()) 
are_collinear(x1,y1,x2,y2,x3,y3)
```

## W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
n = list(map(int,input("nhập n số nguyên:").split()))
def count(n):
    for i in n:
        if i < 0:
            n[n.index(i)] = -i
        else:
            continue
    dem = 0
    for i in n:
        i = str(i)
        for k in i:
            if k != "0":
                dem += 1
            else:
                continue
    return dem
count(n)
```

## W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
def min_max(lst):
    Min = lst[0]
    Max = lst[0]
    for i in lst:
        if i > Max:
            Max = i
        elif i < Min:
            Min = i
        else:
            continue
    return Max,Min
lst = list(map(int,input("nhập 3 số nguyên:").split()))
min_max(lst)
```

## W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n

```python
def sum_of_squares(n):
    tong = 0
    for i in range(1,n+1):
        tong += i**2
    return tong
sum_of_squares(int(input("nhập n:")))
```
